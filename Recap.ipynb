# Conditional Statements

`if / elif / else` lets your code branch based on a condition. Python uses **indentation** (not braces `{}`)
to define each block — this is the single most common source of beginner errors.

```python
age = 20

if age < 12:
    category = "Child"
elif age < 18:
    category = "Teen"
elif age < 60:
    category = "Adult"
else:
    category = "Senior"

print("===== CONDITIONAL STATEMENTS =====")
print(f"Age {age} is classified as: {category}")
print("------------------------")

# Combining with logical operators
fare = 8.05
qualifies_for_discount = (age < 12) or (fare < 5)
print("Qualifies for discount:", qualifies_for_discount)
```

**Explanation:** Python checks conditions top to bottom and stops at the **first** `True` branch — later `elif`
branches are skipped even if they'd also be true. Combine conditions with `and` / `or` / `not` for more complex logic.

**Mini Exercise:** Change `age` to `15` and confirm the category becomes `"Teen"`. Then test `age = 65`.

```python
# Mini Exercise: conditionals
for test_age in [15, 65]:
    if test_age < 12:
        result = "Child"
    elif test_age < 18:
        result = "Teen"
    elif test_age < 60:
        result = "Adult"
    else:
        result = "Senior"
    print(test_age, "->", result)
```

# Loops

`for` loops iterate over a sequence (list, string, range...). `while` loops repeat as long as a condition
stays `True`. `break` exits a loop early; `continue` skips to the next iteration.

```python
print("===== FOR LOOP =====")
for number in range(1, 6):
    print(number)
print("------------------------")

print("===== WHILE LOOP =====")
count = 1
while count <= 5:
    print(count)
    count += 1
print("------------------------")

print("===== BREAK & CONTINUE =====")
for number in range(1, 10):
    if number == 4:
        continue   # skip 4
    if number == 7:
        break      # stop at 7
    print(number)
```

**Explanation:** `range(1, 6)` generates `1, 2, 3, 4, 5` (stops before 6). `while` loops need a variable that
changes (`count += 1`) or they'll run forever — always double-check your loop has a way to stop.

# Dictionaries

```python
print("===== DICTIONARY =====")

student = {
    "name": "Ali",
    "age": 20,
    "courses": ["Python", "Data Science"]
}

print("Original:", student)
print("Name:", student["name"])

student["grade"] = "A"
print("After Add:", student)

student["age"] = 21
print("After Update:", student)

del student["grade"]
print("After Remove:", student)

student.pop("courses")
print("After Remove2:", student)
print("------------------------")
```

```python
print("===== LOOPING THROUGH DICTIONARIES =====")

student = {
    "name": "Ali",
    "age": 20,
    "grade": "A"
}

print("Loop through KEYS:")
for key in student.keys():
    print(key)
print("------------------------")

print("Loop through VALUES:")
for value in student.values():
    print(value)
print("------------------------")

print("Loop through ITEMS (key-value pairs):")
for key, value in student.items():
    print(key, ":", value)
print("------------------------")
```

# List Comprehension

A compact way to build a list from an existing iterable: `[expression for item in iterable if condition]`.

```python
# Traditional loop
squares_traditional = []
for n in range(1, 6):
    squares_traditional.append(n ** 2)

# Equivalent list comprehension
squares_comprehension = [n ** 2 for n in range(1, 6)]

print("===== LIST COMPREHENSION =====")
print(squares_traditional)
print(squares_comprehension)
print("------------------------")

# With a condition
even_numbers = [n for n in range(1, 21) if n % 2 == 0]
print("Even numbers:", even_numbers)
```

**Explanation:** Both approaches produce identical results — list comprehension is just more compact. You'll
see this style constantly in real Pandas and data-cleaning code, so recognizing it matters even before you use
it fluently yourself.

**Mini Exercise:** Write a list comprehension that returns the square of only the odd numbers from 1 to 10.

# Lambda Functions (Brief Introduction)

A `lambda` is a small, anonymous, single-expression function — often used inline with functions like `sorted()`.

```python
students_scores = [("Kate", 88), ("John", 72), ("Amir", 95)]

# Named function
def get_score(student):
    return student[1]

sorted_named = sorted(students_scores, key=get_score)

# Equivalent lambda
sorted_lambda = sorted(students_scores, key=lambda student: student[1])

print("===== LAMBDA =====")
print(sorted_named)
print(sorted_lambda)
```

**Explanation:** `lambda student: student[1]` is a tiny unnamed function that tells `sorted()` to sort by the
second item (the score) of each tuple — identical result to writing a full `def` function, in one line.
This is a preview only; deeper `lambda` usage returns with Pandas `.apply()` in Lecture 3.

**Mini Exercise:** Sort `students_scores` in descending order using `reverse=True`.

```python
# Mini Exercise: lambda
sorted_desc = sorted(students_scores, key=lambda student: student[1], reverse=True)
print(sorted_desc)
```

**Mini Exercise:** Write a function `calculate_max(numbers)` that returns the highest value in a list, without using Python's built-in `max()`.

```python
# Mini Exercise: functions
def calculate_max(numbers):
    highest = numbers[0]
    for n in numbers:
        if n > highest:
            highest = n
    return highest

print(calculate_max([4, 19, 7, 2]))
```

# Variable Scope

```python
print("===== VARIABLE SCOPE =====")

x = 10  # Global variable

def show():
    print("Accessing global x inside function:", x)
    y = 5  # Local variable
    print("Local y:", y)

show()

def modify_global():
    global x
    x = 7
    print(x)

modify_global()
print(x)
```

**Explanation:** A variable created inside a function (`y`) only exists inside that function — this is called *local scope*. Functions can *read* global variables freely, but need the `global` keyword to *modify* them.

# Common Mistakes

Every beginner hits these three error types. Below, each one is triggered **for real** and safely caught with
`try/except` so you can see the actual error message without stopping the notebook.

```python
print("===== SYNTAX ERROR EXAMPLE =====")

broken_code = '''
if True:
print("Hello")
'''

try:
    exec(broken_code)
except SyntaxError as e:
    print("SyntaxError caught:", e)
    print("Fix: indent the line under 'if True:' with 4 spaces.")
```

```python
print("===== RUNTIME ERROR EXAMPLES =====")

# 1) Division by zero
try:
    x = 10
    y = 0
    print(x / y)
except ZeroDivisionError as e:
    print("ZeroDivisionError:", e)

# 2) Index out of range
try:
    nums = [1, 2, 3]
    print(nums[5])
except IndexError as e:
    print("IndexError:", e)

# 3) Using an undefined variable
try:
    print(total_score)
except NameError as e:
    print("NameError:", e)
```

```python
print("===== LOGICAL ERROR EXAMPLE =====")

age = 17

# Bug: this excludes 18 and 19 year-olds from "Adult" by mistake
if age > 20:
    print("Adult")
else:
    print("Not Adult")

# The code runs with no error message at all — but the logic is wrong.
# A 17-year-old correctly prints "Not Adult" here, but so would an 18 or 19-year-old, which is incorrect.
```

**Explanation:**
- A **Syntax Error** means Python can't even understand your code's structure (here: missing indentation).
- A **Runtime Error** means the code is structurally valid but fails while running (dividing by zero, accessing
  an index that doesn't exist, referencing a variable that was never created).
- A **Logical Error** is the most dangerous type: the code runs perfectly with no error message, but produces
  the *wrong* answer because the logic itself is flawed. Always test your conditions with edge-case values.

## Practice Exercise

Combine what you learned in this notebook:
1. Write a function `average_column(df, column_name)` that returns the mean of a numeric column, rounded to 2 decimals.
2. Use it to print the average `Age` and average `Fare` from `titanic_df`.
3. Using a **list comprehension**, build a list of unique values in the `Pclass` column.
4. Using an `if/elif/else`, classify the average fare as `"Low"` (< 15), `"Medium"` (15–50), or `"High"` (> 50).

```python
def average_column(df, column_name):
    '''Return the mean of a numeric column, rounded to 2 decimals.'''
    return round(df[column_name].mean(), 2)

avg_age = average_column(titanic_df, "Age")
avg_fare = average_column(titanic_df, "Fare")
print("Average Age:", avg_age)
print("Average Fare:", avg_fare)

unique_classes = list({p for p in titanic_df["Pclass"]})
print("Unique Passenger Classes:", unique_classes)

if avg_fare < 15:
    fare_level = "Low"
elif avg_fare <= 50:
    fare_level = "Medium"
else:
    fare_level = "High"

print("Average Fare Level:", fare_level)
```

## Challenge (Optional — for students who finish early)

Write a function `summarize_dataset(df)` that:
- Prints the total number of rows and columns.
- Counts survivors vs. non-survivors using a `for` loop and `if/else` (not a shortcut pandas method).
- Uses a list comprehension to list all `Fare` values above the dataset's average fare.

```python
def summarize_dataset(df):
    '''Print a simple summary of the dataset using only this notebook's tools.'''
    rows, cols = df.shape
    print(f"Rows: {rows}, Columns: {cols}")

    survived_count = 0
    not_survived_count = 0
    for value in df["Survived"]:
        if value == 1:
            survived_count += 1
        else:
            not_survived_count += 1
    print(f"Survived: {survived_count}, Did Not Survive: {not_survived_count}")

    avg_fare = df["Fare"].mean()
    above_average_fares = [f for f in df["Fare"] if f > avg_fare]
    print(f"Passengers who paid above the average fare: {len(above_average_fares)}")

summarize_dataset(titanic_df)
```
